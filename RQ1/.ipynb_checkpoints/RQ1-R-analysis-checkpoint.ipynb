```r
library(ggplot2)
library(dplyr)
```

```r
#Violin plot of code size
df= read.csv("RQ1-Relative-frequency.csv")
summary(df)
```

```r
ggplot(df, aes(fill=is_dac,x=as.factor(is_dac), y=avg_code_size))+
scale_y_continuous(trans='log2')+
geom_violin(position = "dodge",draw_quantiles=c(0.25, 0.5, 0.75)) +
xlab("Types of class") + 
ylab (expression(paste("Source lines of code (average)"))) +
scale_fill_discrete(name = "Class type")+
theme(legend.position = "bottom")
ggsave("RQ1_loc_dac_reg.pdf",width=6, height = 4, units = "in")
```

```r
#plot for relative frequency
ggplot(df, aes(fill=is_dac,x=as.factor(is_dac), y=Relative_frequency))+
scale_y_continuous(trans='log2')+
geom_violin(position = "dodge",draw_quantiles=c(0.25, 0.5, 0.75)) +
xlab("Types of classes") + 
ylab (expression(paste("Refactoring density"))) +
scale_fill_discrete(name = "Class types")+
theme(legend.position = "bottom")
ggsave("RQ1_RF_dac_reg.pdf",width=6, height = 4, units = "in")
```

```r
#compute the median and mean
dac<-subset(df, is_dac == "DAC")
reg<-subset(df, is_dac == "Regular")
```

```r
summary(dac)
summary(reg)
reg_sampled<- sample_n(reg, 1318)
```

```r
summary(reg_sampled)
```

```r
#perform statiscial test
#Run Wilcoxon rank sum test or Mann-Whitney test
wilcox.test(reg_sampled$Relative_frequency, dac$Relative_frequency,  conf.int = TRUE)
```
